```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold, GridSearchCV, cross_val_score, cross_val_predict
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('abalone.csv')
data['Sex'] = data['Sex'].map(lambda x: 1 if x == 'M' else (-1 if x == 'F' else 0))
#print(data)
X = np.array(data.iloc[:,0:-1]) # convert DataFrame to Array
y = np.array(data.iloc[:,-1]) # convert DataFrame to Array
```

```python
data
```

```python
#Variant N1

# Create the parameter grid based on the results of random search 
param_grid = {
    #'bootstrap': [True],
    #'max_depth': [80, 90, 100, 110],
    #'max_features': [2, 3],
    #'min_samples_leaf': [3, 4, 5],
    #'min_samples_split': [8, 10, 12],
    'random_state': [1],
    'n_estimators': np.arange(1,51)
    #'n_estimators': [10, 20, 30, 50]
}
# Create a based model
rf = RandomForestRegressor()
#regr = RandomForestRegressor(random_state=1, n_estimators=p)
kf = KFold(n_splits=5, random_state=1, shuffle=True)
gs = GridSearchCV(estimator = rf, param_grid = param_grid, scoring='r2', cv=kf, return_train_score=True)
gs.fit(X,y)
print(gs.best_params_)
cv_results = pd.DataFrame(gs.cv_results_) # http://qaru.site/questions/16902941/what-is-the-alternative-function-for-meanvalidationscore-in-gridcvresults
c_o1 = np.max(cv_results.mean_test_score) # max value score
print(c_o1)
print(cv_results.mean_test_score)
```

```python
#print(np.argmax(cv_results.mean_test_score))

f= open("answers_12/1.txt","w+")
#f.write(str(np.round(corr,decimals=2)))
#r = np.round(cv_results.mean_test_score,decimals=2)
nr = np.min(np.where(cv_results.mean_test_score > 0.52))+1+1 # +1 it's misunderstanding!!!
f.write(str(nr))
f.close()
print(nr)

```

```python
#Variant N2 (from forum)
kf = KFold(n_splits=5, shuffle=True, random_state=1)
for i in range (1, 51):
    clf = RandomForestRegressor(random_state=1, n_estimators=i)
    sc = cross_val_score(estimator=clf, cv=kf, X=X, y=y, scoring = "r2")
    #print(i)
    if np.mean(sc) > 0.52:
        print(i)
        break
```

```python
# for forum
from sklearn.model_selection import KFold, GridSearchCV
param_grid = {
    'random_state': [1],
    'n_estimators': np.arange(1,51)
}
rf = RandomForestRegressor()
kf = KFold(n_splits=5, random_state=1, shuffle=True)
gs = GridSearchCV(estimator = rf, param_grid = param_grid, scoring='r2', cv=kf, return_train_score=True)
gs.fit(X,y)
cv_results = pd.DataFrame(gs.cv_results_)
print(np.min(np.where(cv_results.mean_test_score > 0.52))+1)
```
